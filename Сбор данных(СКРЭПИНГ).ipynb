```python
import pandas as pd
import numpy as np
import datetime
import time

from bs4 import BeautifulSoup
import requests
import re
import csv

import plotly.express as px
```

```python

page_book_url = [] # page_book_url - записать ссылки на все книги

def main():
    url = 'https://www.flip.kz/'
    pages = page_urls()
    for page_url in pages:
        book_url = book_urls(page_url)
        for url in book_url:
            xyz = 'https://www.flip.kz/' + url
            page_book_url.append(xyz)

def book_urls(page_url):
    html = get_html(page_url)
    href_page = get_book_page(html)
    
    return href_page
    

def get_html(url):
    responce = requests.get(url)
    return responce.text


def get_book_page(html):
    book_href = []
    soup = BeautifulSoup(html, 'lxml')
    books = soup.find_all('div', class_='info')
    for book in books:
        hre = book.find('a', class_='title').get('href')
        book_href.append(hre)
    
    return book_href
    
def page_urls():
    links = []
    for i in range(1, 51):
        page_url = 'https://www.flip.kz/catalog?subsection=1&filter-show=1&page=' + str(i)
        links.append(page_url)
    return links

    
if __name__ == '__main__':
    main()




```

```python
df = []


def main():
    for index, book in enumerate(page_book_url): 
        print(book, f'   {index+1}')
        html = get_html_book(book)
        data = get_data_book(html)
        

def get_html_book(url):
    responce = requests.get(url)
    return responce.text


def get_data_book(html):
    soup = BeautifulSoup(html, 'lxml')
    try:
        name = soup.find('span', itemprop="name").text
    except:
        name = ''
    try:
        price = ''.join(re.findall(r'[0-9]' ,(soup.find('span', class_='text_att', style="font-size:120%").text)))
    except:
        price = ''
    try:
        author = soup.find('p', style='margin-bottom: 0px;').text
    except:
        author = ''
    try:
        discount_price = ''.join(re.findall(r'[0-9]', (soup.find('span', class_="text_att", style="font-size:120%").text)))
    except:
        discount_price = ''
    try:
        number_of_reviews = soup.find('span', itemprop="reviewCount").text
    except:
        number_of_reviews = ''
    try:
        rating = soup.find('span', itemprop="ratingValue").text
    except:
        rating = ''
    try:
        publishing_house = soup.find('div', class_='description-table').find('a').text
    except:
        publishing_house = ''
    try:
        year_of_publication = ''.join(re.findall(r'[0-9]' ,(soup.find('div', class_='description-table').find('div', class_='description-cell').next_element.next_element.text)))
    except:
        year_of_publication = ''
    try:
        bookbinding = (soup.find_all('div', class_='description-row')[1].next_element.next_element.next_element.next_element).split(',')[0].strip()
    except:
        bookbinding = ''
    try:
        quantity_of_pages = ''.join(re.findall(r'[0-9]', (soup.find_all('div', class_='description-row')[1].next_element.next_element.next_element.next_element).split(',')[1].strip()))
    except:
        quantity_of_pages = ''
    try:
        language = soup.find_all('div', class_='description-row')[-2].next_element.next_element.next_element.next_element
    except:
        language = ''
    try:
        quantity_of_ratings_5 = soup.find_all('div', class_='row')[0].find('div', class_='cell w-0 font_small').text
    except:
        quantity_of_ratings_5 = ''
    try:
        quantity_of_ratings_4 = soup.find_all('div', class_='row')[1].find('div', class_='cell w-0 font_small').text
    except:
        quantity_of_ratings_4 = ''
    try:
        quantity_of_ratings_3 = soup.find_all('div', class_='row')[2].find('div', class_='cell w-0 font_small').text
    except:
        quantity_of_ratings_3 = ''
    try:
        quantity_of_ratings_2 = soup.find_all('div', class_='row')[3].find('div', class_='cell w-0 font_small').text
    except:
        quantity_of_ratings_2 = ''
    try:
        quantity_of_ratings_1 = soup.find_all('div', class_='row')[4].find('div', class_='cell w-0 font_small').text
    except:
        quantity_of_ratings_1 = ''

    
    df.append({'book_name': name, 'author': author, 'price(тг)': price, 'discount_price(тг)': discount_price,
               'number_of_reviews': number_of_reviews, 'rating(5)': rating, 'publishing_house': publishing_house,
               'year_of_publication': year_of_publication, 'bookbinding': bookbinding,
               'quantity_of_pages': quantity_of_pages, 'language': language, 'quantity_of_ratings_5': quantity_of_ratings_5,
               'quantity_of_ratings_4': quantity_of_ratings_4, 'quantity_of_ratings_3': quantity_of_ratings_3,
               'quantity_of_ratings_2': quantity_of_ratings_2, 'quantity_of_ratings_1': quantity_of_ratings_1})
    
    return 0


if __name__ == '__main__':
    main()
    
df_flip = pd.DataFrame(df)
```
